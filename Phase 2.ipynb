###### convert FL_DATE to Year, Month, Day_of_Week

```python
# Cleaning 

# 1. Drop redundant columns
drop_cols = ['AIRLINE', 'AIRLINE_DOT', 'DOT_CODE', 'ORIGIN_CITY', 'DEST_CITY']
df_clean = df.drop(columns=drop_cols)

# 2. Convert FL_DATE to datetime
df_clean['FL_DATE'] = pd.to_datetime(df_clean['FL_DATE'], errors='coerce')

# Extract year, month, day of week (0=Mon)
df_clean['YEAR'] = df_clean['FL_DATE'].dt.year
df_clean['MONTH'] = df_clean['FL_DATE'].dt.month
df_clean['DAY_OF_WEEK'] = df_clean['FL_DATE'].dt.dayofweek

# 3. Replace NaN in delay cause columns with 0 AKA no delay
delay_cols = [
    'DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER', 'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY', 'DELAY_DUE_LATE_AIRCRAFT'
]
df_clean[delay_cols] = df_clean[delay_cols].fillna(0)

print("New shape after cleaning:", df_clean.shape)
print("\nRemaining columns:", df_clean.columns.tolist()[:31], "...")
```

```python
# Convert HHMM time columns to hour/minute features 

import numpy as np
def extract_hour_minute(t):
    """Convert time in HHMM (int/float) to hour and minute."""
    if pd.isna(t):
        return np.nan, np.nan
    t = int(t)
    hour = t // 100
    minute = t % 100
    # Handle invalid times (e.g. 2400)
    if hour >= 24 or minute >= 60:
        return np.nan, np.nan
    return hour, minute

# Time columns to process
time_cols = ['CRS_DEP_TIME', 'DEP_TIME', 'CRS_ARR_TIME', 'ARR_TIME']

for col in time_cols:
    df_clean[f"{col}_HOUR"], df_clean[f"{col}_MINUTE"] = zip(*df_clean[col].map(extract_hour_minute))
df_clean = df_clean.drop(columns=time_cols)

print("✅ Time conversion complete.")
print("Remaining columns:", df_clean.columns.tolist()[:15], "...")
```

```python
df_clean[['CRS_DEP_TIME_HOUR', 'CRS_DEP_TIME_MINUTE', 'DEP_TIME_HOUR', 'DEP_TIME_MINUTE']].head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create delay classification label
df_clean['IS_DELAYED'] = (df_clean['ARR_DELAY'] > 15).astype(int)

# Rename CANCELLED to a clearer label name
df_clean.rename(columns={'CANCELLED': 'IS_CANCELLED'}, inplace=True)

# Handle CANCELLATION_CODE missing values (fill with 'None')
df_clean['CANCELLATION_CODE'] = df_clean['CANCELLATION_CODE'].fillna('None')

# Show proportions
print("Delay distribution (%):")
print(df_clean['IS_DELAYED'].value_counts(normalize=True) * 100)

print("\nCancellation distribution (%):")
print(df_clean['IS_CANCELLED'].value_counts(normalize=True) * 100)

# Show sample of cancellation codes
print("\nCancellation code sample:")
print(df_clean['CANCELLATION_CODE'].value_counts())
```

```python
# Categorical columns to encode
cat_cols = ['AIRLINE_CODE', 'ORIGIN', 'DEST', 'CANCELLATION_CODE']

# Initialize encoders dictionary to keep mappings
encoders = {}

# Apply label encoding
for col in cat_cols:
    le = LabelEncoder()
    df_clean[col] = le.fit_transform(df_clean[col])
    encoders[col] = le  # store encoder if you want to inverse later

print("✅ Encoding complete.")
print(df_clean[cat_cols].head())
```
