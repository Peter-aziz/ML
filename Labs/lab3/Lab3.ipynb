### 2.2.2 Dataset Loading

#### Define method to download & load dataset

```python
def load_dataset():
    download_file("https://raw.githubusercontent.com/KhaledElTahan/DeepLearning/master/Labs/lab3/lab3_train.h5")
    download_file("https://raw.githubusercontent.com/KhaledElTahan/DeepLearning/master/Labs/lab3/lab3_test.h5")

    path_to_train = "lab3_train.h5"
    path_to_test = "lab3_test.h5"

    train_dataset = h5py.File(path_to_train, "r")
    train_x = np.array(train_dataset['train_set_x'][:])
    train_y = np.array(train_dataset['train_set_y'][:])

    test_dataset = h5py.File(path_to_test, "r")
    test_x = np.array(test_dataset['test_set_x'][:])
    test_y = np.array(test_dataset['test_set_y'][:])

    # reshape y from (samples, ) to (1, samples)
    train_y = train_y.reshape((1, train_x.shape[0]))
    test_y = test_y.reshape((1, test_x.shape[0]))

    # transpose y
    train_y = train_y.T
    test_y = test_y.T

    return train_x, train_y, test_x, test_y
```

#### Define method to make simple preprocessing on the dataset

```python
def preprocess_data():
    train_x, train_y, test_x, test_y = load_dataset()

    # Normalize image vectors
    train_x = train_x/255.
    test_x = test_x/255.

    print ("number of training examples = " + str(train_x.shape[0]))
    print ("number of test examples = " + str(test_x.shape[0]))

    print ("X_train shape: " + str(train_x.shape))
    print ("Y_train shape: " + str(train_y.shape))
    print ("X_test shape: " + str(test_x.shape))
    print ("Y_test shape: " + str(test_y.shape))

    return train_x, train_y, test_x, test_y
```

#### Get the preprocessed Data

### 2.2.4 Plot Utility Methods

The following code is used to plot accuracy and loss histories for each model experiment.

```python
def plot(training_results, validation_results, results_type, model_name):
    fig = plt.figure(figsize=[8, 6])

    plt.plot(training_results, 'r', linewidth=3.0)
    plt.plot(validation_results, 'b', linewidth=3.0)
    plt.legend(['Training ' + results_type, 'Validation ' + results_type], fontsize=18)
    plt.xlabel('Epochs', fontsize=16)
    plt.ylabel(results_type, fontsize=16)
    plt.title(results_type + ' of ' + model_name, fontsize=16)


def plot_accuracy(history, model_name):
    plot(history.history['accuracy'], history.history['val_accuracy'], 'Accuracy', model_name)


def plot_loss(history, model_name):
    plot(history.history['loss'], history.history['val_loss'], 'Loss', model_name)
```

```python
def CustomCNN(input_shape):
    """
    Implementation of the Custom CNN.
    Args:
        input_shape () shape of the images of the dataset
    Returns:
        model (Keras.Model): CNN Custom Model
        model_name (Str): The name of the model
    """

    ### START CODE HERE ###
    X_input = Input(input_shape) #''' TODO: Define the input layer '''

    # Zero-Padding: pads the border of X_input with zeroes
    X = ZeroPadding2D((3, 3))(X_input)

    # CONV -> BN -> RELU Block applied to X
    # Choose a suitable number of filters and kernel size
    X = Conv2D(filters=32, kernel_size=(7, 7), strides=(1, 1), name='conv0')(X) #'''TODO: Define the conv2d layer'''
    X = BatchNormalization(axis=3, name='bn0')(X)
    X = Activation('relu')(X) #'''TODO: Define the Relu activation layer'''

    # MAXPOOL
    X = MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='valid')(X) #'''TODO: Define the maxpooling2d layer'''
    # FLATTEN X (means convert it to a vector) + FULLYCONNECTED (Dense)
    X = Flatten()(X)
    Y = Dense(1, activation='sigmoid')(X) #'''TODO: Define the output dense layer'''

    # Create model. This creates your tensorflow.keras model instance, you'll use this instance to train/test the model.
    model = Model(inputs=X_input, outputs=Y)
    ### END CODE HERE ###

    return model, "Custom CNN"
```

```python
model, model_name = CustomCNN((64, 64, 3))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(train_x, train_y, epochs=20, validation_split=0.1)
```

```python
# Plot accuracy and loss
plot_accuracy(history, model_name)
plot_loss(history, model_name)

# Evaluate on test set
preds = model.evaluate(test_x, test_y)
print("Test Loss =", preds[0])
print("Test Accuracy =", preds[1])
```

```python
def VGG(pretrained=True, input_shape=(64, 64, 3), model_name="VGG16", freeze_layers=0):
    """
    Returns VGG Keras Model
    Args:
        pretrained (Bool): Whether get a pretrained model or not
        input_shape (Tuple): To create the input layer, should match the image resolution
        model_name (str): String to capture the model name
    Examples:
        VGG(True) -> Model, "Pretrained VGG16"
    """
    # Select VGG16 or VGG19
    if model_name == "VGG16":
        vgg_app = applications.vgg16.VGG16
    else:
        vgg_app = applications.vgg19.VGG19

    # Load with or without pretrained weights
    if(pretrained):
        full_model_name = "Pretrained " + model_name
        base_model = vgg_app(weights='imagenet', include_top=False, input_shape=input_shape, pooling='none')
    else:
        full_model_name = "Untrained " + model_name
        base_model = vgg_app(weights=None, include_top=False, input_shape=input_shape, pooling='none')

    # Add classification head on top of base model
    X = base_model.output
    X = Flatten()(X)
    X = Dense(256, activation='relu')(X)
    X = Dropout(0.5)(X)
    Y = Dense(1, activation='sigmoid')(X)

    model = Model(inputs=base_model.input, outputs=Y)

    # Freeze AFTER building the full model
    if freeze_layers > 0:
        full_model_name = "Pretrained " + model_name + " (Frozen " + str(freeze_layers) + " layers)"
        for layer in base_model.layers[:freeze_layers]:
            layer.trainable = False

    # Verify freezing is working
    #print(f"\nModel: {full_model_name}")
    #for i, layer in enumerate(model.layers):
    #    print(f"  Layer {i}: {layer.name} | Trainable: {layer.trainable}")


    return model, full_model_name
```

```python
def train_and_evaluate(model, model_name, train_x, train_y, test_x, test_y, epochs=20, lr=1e-4):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    history = model.fit(train_x, train_y, epochs=epochs, batch_size=16, validation_split=0.1)

    plot_accuracy(history, model_name)
    plot_loss(history, model_name)

    print("\n--- Evaluation ---")
    train_loss, train_acc = model.evaluate(train_x, train_y)
    print(f"Train Accuracy: {train_acc:.4f}")

    test_loss, test_acc = model.evaluate(test_x, test_y)
    print(f"Test Accuracy:  {test_acc:.4f}")

    return history, test_acc
```

```python
# The ResNet function refers to the above residual connection-based networks including the InceptionResNetV2 network
def ResNet(pretrained = True, input_shape = (64, 64, 3), model_name = "ResNet50", freeze_layers=0):
    """
    Returns ResNet Keras Model
    Args:
        pretrained (Bool): Whether get a pretrained model or not
        input_shape (Tuple): To create the input layer, should match the image resolution
        model_name (str): String to capture the model name
    Examples:
        ResNet(True) -> Model, "Pretrained ResNet50"
    """

    # Select the correct architecture
    if model_name == "ResNet50":
        resnet_app = applications.resnet50.ResNet50
    elif model_name == "ResNet101":
        resnet_app = applications.resnet.ResNet101
    else:  # InceptionResNetV2
        resnet_app = applications.inception_resnet_v2.InceptionResNetV2


    if(pretrained):
        full_model_name  = "Pretrained " + model_name
        base_model = resnet_app(weights='imagenet', include_top=False, input_shape=input_shape, pooling='none')
    else:
        full_model_name  = "Untrained " + model_name
        base_model = resnet_app(weights=None, include_top=False, input_shape=input_shape, pooling='none')

    # classification head
    X = base_model.output
    X = Flatten()(X)
    X = Dense(256, activation='relu')(X)
    X = Dropout(0.5)(X)
    Y = Dense(1, activation='sigmoid')(X)

    model = Model(inputs=base_model.input, outputs=Y)

    # Freeze AFTER building full model
    if freeze_layers > 0:
        full_model_name = "Pretrained " + model_name + " (Frozen " + str(freeze_layers) + " layers)"
        for layer in model.layers[:freeze_layers]:
            layer.trainable = False


    return model, full_model_name
```

```python
def train_and_evaluate(model, model_name, train_x, train_y, test_x, test_y, epochs=20, lr=1e-4):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    history = model.fit(train_x, train_y, epochs=epochs, batch_size=16, validation_split=0.1)

    plot_accuracy(history, model_name)
    plot_loss(history, model_name)

    print("\n--- Evaluation ---")
    train_loss, train_acc = model.evaluate(train_x, train_y)
    print(f"Train Accuracy: {train_acc:.4f}")

    test_loss, test_acc = model.evaluate(test_x, test_y)
    print(f"Test Accuracy:  {test_acc:.4f}")

    return history, test_acc
```

### 2.2.9 Layers Freezing

The following utility method is used for layers freezing.

**TODO**: *Experiment all the pretrained models with layers freezing using* **different number of frozen layers**.

**NOTE**: You will not change this method at all, the experimenting will take place in the test cases section below.

```python
def freeze(model, number_of_frozen_layers):
    layers = model.layers

    layers = layers[:number_of_frozen_layers]

    for layer in layers:
        layer.trainable = False

    return model
```
